```python
import pandas as pd
```

```python
url = 'datasets/pesquisa_musica.csv'
df_musica = pd.read_csv(url)
df_musica
```

```python
df_musicas = df_musica.drop('Carimbo de data/hora', axis=1)
```

```python
df_musicas.columns = ['rock','hip-hop','mpb','reggae','funk','sertanejo','gospel','eletronica']
```

```python
df_musicas.head()
```

```python
import matplotlib.pyplot as plt

media_estilos = df_musicas.mean()

# Cria o gr谩fico de barras
media_estilos.plot(kind='bar', figsize=(12, 8), color='skyblue', alpha=0.75)

# Adiciona t铆tulo e r贸tulos aos eixos
plt.title('M茅dia das Notas por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Nota M茅dia')

# Adiciona r贸tulos s barras
for i, valor in enumerate(media_estilos):
    plt.text(i, valor + 0.1, f'{valor:.2f}', ha='center', va='bottom')

# Exibe o gr谩fico
plt.tight_layout()  # Ajusta o layout para n茫o cortar o gr谩fico
plt.show()
```

```python
import seaborn as sns

# Cria o boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_musica, palette='Set2')

# Adiciona t铆tulo e r贸tulos aos eixos
plt.title('Boxplot das Notas por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Nota')

# Exibe o gr谩fico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cria um histograma para cada estilo musical
plt.figure(figsize=(14, 12))

for i, column in enumerate(df_musicas.columns, 1):
    plt.subplot(3, 3, i)  # Ajusta o layout para v谩rios subgr谩ficos
    sns.histplot(df_musica[column], bins=5, kde=True, color='skyblue')
    plt.title(f'Estilo: {column}')
    plt.xlabel('Nota')
    plt.ylabel('Frequ锚ncia')

plt.tight_layout()
plt.show()
```
