```python
import pandas as pd
```

```python
url = 'datasets/pesquisa_turma.csv'
df_turma = pd.read_csv(url)
df_turma
```

```python
df_turma.columns = ['data_hora','email','pontuacao','compreensao','desafiador','sentimento']
df_turma
```

```python
import seaborn as sns

# Criar o DataFrame com a contagem de cada sentimento
df = df_turma['sentimento'].value_counts(normalize=True).reset_index()
df.columns = ['sentimento', 'porcentagem']

# Criar o gráfico de barras
ax = sns.barplot(x='sentimento', y='porcentagem', data=df)

# Adicionar as porcentagens nas barras
ax.bar_label(ax.containers[0], fmt='%.2f%%')

# Mostrar o gráfico
sns.boxplot()
```
