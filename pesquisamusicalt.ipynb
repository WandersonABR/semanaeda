# Importando pandas

```python
import pandas as pd
```

```python
url = 'datasets/PesquisaSimulada.csv'
df_musicas = pd.read_csv(url)
```

```python
df_musicas.head()
```

```python
df_musica = df_musicas.drop('Carimbo de data/hora', axis=1)
```

```python
df_musica.columns = ['rock', 'hip-hop', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica']
```

```python
df_musica.head()
```

```python
df_notas = df_musicas.sum()
df_notas
```

# Média das notas por estilo musical

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = df_musica
df = pd.DataFrame(df_musica)
mean_scores = df.mean()

# Criar gráfico de barras
plt.figure(figsize=(10, 6))
ax = mean_scores.plot(kind='bar', color='skyblue')
plt.title('Média das Notas por Estilo Musical')
plt.xlabel('Estilo Musical')
plt.ylabel('Nota Média')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Adicionar valores no gráfico
for i, valor in enumerate(mean_scores):
    plt.text(i, valor + 0.1, f'{valor:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Calcular a média das notas para cada estilo musical
mean_scores = df.mean()

# Criar gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(mean_scores.index, mean_scores.values, marker='o', linestyle='-', color='b')
plt.title('Média das Notas por Estilo Musical')
plt.xlabel('Estilo Musical')
plt.ylabel('Nota Média')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Criar o gráfico de boxplot
plt.figure(figsize=(12, 8))

# Configurações do boxplot
boxprops = dict(facecolor='lightblue', color='black')
whiskerprops = dict(color='gray')
medianprops = dict(color='red')

# Criar o boxplot com configurações personalizadas
bp = plt.boxplot(
    df.values,
    vert=True,
    patch_artist=True,
    boxprops=boxprops,
    whiskerprops=whiskerprops,
    medianprops=medianprops,
    capprops=dict(color='black')
)

# Configurar os rótulos
plt.xticks(range(1, len(df.columns) + 1), df.columns)
plt.title('Distribuição das Notas por Estilo Musical')
plt.xlabel('Estilo Musical')
plt.ylabel('Nota')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Cria um histograma para cada estilo musical
plt.figure(figsize=(14, 12))

for i, column in enumerate(df_musica.columns, 1):
    plt.subplot(3, 3, i)  # Ajusta o layout para vários subgráficos
    sns.histplot(df_musica[column], bins=5, kde=True, color='green')
    plt.title(f'Estilo: {column}')
    plt.xlabel('Nota')
    plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
